```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import (
    LabelEncoder,
    StandardScaler
)

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier
```

```python
file_path = '/kaggle/input/datasets/atifkhan12/global-air-pollution-dataset-2025-2026/Global_Air_Pollution_Data_2025_2026.csv'

df = pd.read_csv(file_path)

df.head()
```

```python
X = df.drop('AQI_Class', axis=1)

y = df['AQI_Class']
```

```python
X['Date'] = pd.to_datetime(X['Date'])
```

```python
X['Year'] = X['Date'].dt.year

X['Month'] = X['Date'].dt.month

X['Day'] = X['Date'].dt.day

X['Hour'] = X['Date'].dt.hour
```

```python
if 'Date' in X.columns:
    X = X.drop('Date', axis=1)
```

```python
city_encoder = LabelEncoder()

X['City'] = city_encoder.fit_transform(X['City'])
```

```python
target_encoder = LabelEncoder()

y = target_encoder.fit_transform(y)
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
print(X_train.shape)

print(X_test.shape)

print(y_train.shape)

print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier

from sklearn.ensemble import RandomForestClassifier

from sklearn.neighbors import KNeighborsClassifier

from sklearn.svm import SVC

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)
```

```python
models = {

    'Logistic Regression':
        LogisticRegression(max_iter=5000),

    'Decision Tree':
        DecisionTreeClassifier(),

    'Random Forest':
        RandomForestClassifier(),

    'KNN':
        KNeighborsClassifier(),

    'SVM':
        SVC()

}
```

```python
results = []
```

```python
for name, model in models.items():

    print(f'Training {name}...')

    # Train model
    model.fit(X_train, y_train)

    # Predict
    y_pred = model.predict(X_test)

    # Metrics
    accuracy = accuracy_score(y_test, y_pred)

    precision = precision_score(
        y_test,
        y_pred,
        average='weighted'
    )

    recall = recall_score(
        y_test,
        y_pred,
        average='weighted'
    )

    f1 = f1_score(
        y_test,
        y_pred,
        average='weighted'
    )

    # Store results
    results.append([
        name,
        accuracy,
        precision,
        recall,
        f1
    ])
```

```python
results_df = pd.DataFrame(

    results,

    columns=[
        'Model',
        'Accuracy',
        'Precision',
        'Recall',
        'F1 Score'
    ]
)

results_df
```

```python
results_df.to_csv(
    'rq5_metric_sensitivity.csv',
    index=False
)
```

```python
results_melted = results_df.melt(

    id_vars='Model',

    value_vars=[
        'Accuracy',
        'Precision',
        'Recall',
        'F1 Score'
    ],

    var_name='Metric',

    value_name='Score'
)
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    data=results_melted,
    x='Model',
    y='Score',
    hue='Metric'
)

plt.title(
    'Sensitivity of Model Rankings to Evaluation Metrics'
)

plt.xticks(rotation=15)

plt.savefig(
    'rq5_metric_sensitivity.pdf',
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```
