```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import (
    LabelEncoder,
    StandardScaler
)

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier
```

```python
file_path = '/kaggle/input/datasets/atifkhan12/global-air-pollution-dataset-2025-2026/Global_Air_Pollution_Data_2025_2026.csv'

df = pd.read_csv(file_path)

df.head()
```

```python
X = df.drop('AQI_Class', axis=1)

y = df['AQI_Class']
```

```python
X['Date'] = pd.to_datetime(X['Date'])
```

```python
X['Year'] = X['Date'].dt.year

X['Month'] = X['Date'].dt.month

X['Day'] = X['Date'].dt.day

X['Hour'] = X['Date'].dt.hour
```

```python
if 'Date' in X.columns:
    X = X.drop('Date', axis=1)
```

```python
city_encoder = LabelEncoder()

X['City'] = city_encoder.fit_transform(X['City'])
```

```python
target_encoder = LabelEncoder()

y = target_encoder.fit_transform(y)
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
print(X_train.shape)

print(X_test.shape)

print(y_train.shape)

print(y_test.shape)
```

```python
model = RandomForestClassifier()

model.fit(X_train, y_train)
```

```python
importance = model.feature_importances_
```

```python
importance_df = pd.DataFrame({

    'Feature': X.columns,

    'Importance': importance

})
```

```python
importance_df = importance_df.sort_values(
    by='Importance',
    ascending=False
)

importance_df
```

```python
importance_df.to_csv(
    'rq4_feature_importance.csv',
    index=False
)
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    data=importance_df,
    x='Importance',
    y='Feature'
)

plt.title(
    'Feature Importance for AQI Classification'
)

plt.xlabel('Importance Score')

plt.ylabel('Feature')

plt.savefig(
    'rq4_feature_importance.pdf',
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```
